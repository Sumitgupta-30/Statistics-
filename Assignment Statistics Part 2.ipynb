Theoretical Questions

```python
# Question 1: What is hypothesis testing in statistics?

#Ans) Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. 
#It involves testing a claim or hypothesis about a population parameter using sample data.
```

```python
# Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

#Ans) The null hypothesis (H0) is a statement that there is no effect or no difference in the population, while the alternative hypothesis (H1 or Ha) suggests that there is an effect or a difference.
#     H0: There is no difference or effect.
#     Ha: There is a difference or effect.
```

```python
# Question 3: What is the significance level in hypothesis testing, and why is it important?

#Ans) The significance level (denoted as alpha, α) represents the probability of rejecting the null hypothesis when it is actually true (Type I error).
# It is the threshold for determining whether the observed data is statistically significant. Typically, common significance levels are 0.05, 0.01, or 0.10.

# The significance level is crucial because it helps control the likelihood of Type I errors and ensures valid conclusions.
```

```python
# Question 4: What does a P-value represent in hypothesis testing?

#Ans) The p-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true. 
# It helps determine the strength of the evidence against the null hypothesis. If p-value < significance level, reject the null hypothesis.
```

```python
# Question 5: How do you interpret the P-value in hypothesis testing?

#Ans) If the p-value is less than the significance level (e.g., p < 0.05), we reject the null hypothesis and conclude that the result is statistically significant.
# If the p-value is greater than the significance level (e.g., p > 0.05), we fail to reject the null hypothesis, meaning there is insufficient evidence to support the alternative hypothesis.
```

```python
# Question 6: What are Type 1 and Type 2 errors in hypothesis testing?

#Ans) Type 1 error (False Positive): Rejecting the null hypothesis when it is actually true.
# Type 2 error (False Negative): Failing to reject the null hypothesis when it is actually false.
```

```python
# Question 7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

#Ans) A one-tailed test looks for evidence of an effect in only one direction (either greater than or less than).
# A two-tailed test looks for evidence of an effect in both directions (either greater than or less than).
# For example:

# One-tailed: H0: μ ≤ 10, Ha: μ > 10
# Two-tailed: H0: μ = 10, Ha: μ ≠ 10
```

```python
# Question 8: What is the Z-test, and when is it used in hypothesis testing?

#Ans) The Z-test is used to determine whether there is a significant difference between the sample mean and the population mean (or between two sample means) when the population standard deviation is known.
# It is typically used for large sample sizes (n > 30) and assumes a normal distribution.
```

```python
# Question 9: How do you calculate the Z-score, and what does it represent in hypothesis testing?

#Ans) The Z-score is calculated using the formula:

# Z = (X - μ) / σ
# Where:

# X = sample value
# μ = population mean
# σ = population standard deviation
# The Z-score represents how many standard deviations the sample mean is away from the population mean.
```

```python
# Question 10: What is the T-distribution, and when should it be used instead of the normal distribution?

#Ans) The T-distribution is used when the sample size is small (n ≤ 30) or when the population standard deviation is unknown.
# It has heavier tails compared to the normal distribution, which accounts for the greater uncertainty in small sample sizes.
```

```python
# Question 11: What is the difference between a Z-test and a T-test?

#Ans) The main difference is:

# The Z-test is used when the population standard deviation is known and sample size is large.
# The T-test is used when the population standard deviation is unknown or sample size is small.
# Both tests assess the difference between sample data and population parameters.
```

```python
# Question 12: What is the T-test, and how is it used in hypothesis testing?

#Ans) The T-test is a statistical test used to compare the means of two groups. It is especially useful when the sample size is small and/or the population standard deviation is unknown.

# Types of T-tests:

# One-sample T-test: Compares the sample mean to a known population mean.
# Independent two-sample T-test: Compares the means of two independent groups.
# Paired T-test: Compares the means of two related groups.
```

```python
# Question 13: What is the relationship between Z-test and T-test in hypothesis testing?

#Ans) The Z-test and T-test are both used to compare sample means to population means or between two sample means.
# The key difference is that the Z-test is used when the population standard deviation is known or sample sizes are large, while the T-test is used when the population standard deviation is unknown and sample sizes are small.
```

```python
# Question 14: What is a confidence interval, and how is it used to interpret statistical results?

#Ans) A confidence interval provides a range of values that is likely to contain the population parameter. It is used to estimate the uncertainty of a sample statistic. 
# A 95% confidence interval means that there is a 95% chance that the population parameter lies within this range.
```

```python
# Question 15: What is the margin of error, and how does it affect the confidence interval?

#Ans) The margin of error is the range of values above and below the sample statistic in a confidence interval.
# It represents the uncertainty or variability in the estimate. A larger margin of error results in a wider confidence interval, while a smaller margin results in a narrower interval.
```

```python
# Question 16: How is Bayes' Theorem used in statistics, and what is its significance?

#Ans) Bayes' Theorem is used to calculate the probability of an event, given prior knowledge of conditions that might be related to the event. It is widely used in statistical inference and decision-making. 
# The formula is:

# P(A|B) = (P(B|A) * P(A)) / P(B)
# It is significant because it allows for the updating of probabilities as more evidence or data becomes available.
```

```python
# Question 17: What is the Chi-square distribution, and when is it used?

#Ans) The Chi-square distribution is a continuous probability distribution that is used primarily for categorical data analysis, such as goodness-of-fit tests and tests of independence.
# It is often used when dealing with counts or frequencies in contingency tables.
```

```python
# Question 18: What is the Chi-square goodness of fit test, and how is it applied?

#Ans) The Chi-square goodness of fit test is used to determine whether the observed frequencies of a categorical variable differ significantly from the expected frequencies. 
# It is applied by comparing the observed and expected frequencies and calculating a Chi-square statistic.
```

```python
# Question 19: What is the F-distribution, and when is it used in hypothesis testing?

#Ans) The F-distribution is used to test hypotheses about variances and is the basis for ANOVA (Analysis of Variance).
# It is used when comparing two or more population variances or in regression analysis to test model fit.
```

```python
# Question 20: What is an ANOVA test, and what are its assumptions?

#Ans) ANOVA (Analysis of Variance) is used to compare the means of three or more groups to determine if at least one group mean differs significantly from the others. Assumptions include:

# The populations from which the samples are drawn are normally distributed.
# The variances of the populations are equal (homogeneity of variance).
# The samples are independent.
```

```python
# Question 21: What are the different types of ANOVA tests?

#Ans) The main types of ANOVA are:

# 1. One-way ANOVA: Tests for differences in means across one factor with multiple levels.
# 2. Two-way ANOVA: Tests for differences in means across two factors, considering interaction effects.
# 3. Repeated measures ANOVA: Tests for differences across groups in a repeated measure setting.
```

```python
# Question 22: What is the F-test, and how does it relate to hypothesis testing?

#Ans) The F-test is used to compare two variances to determine if they are significantly different.
# It is commonly used in ANOVA to test whether the variance in sample data comes from the same population.
```

```python
#1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy import stats

# Sample data
sample_data = [15, 16, 17, 14, 13, 19, 20, 18, 17, 16]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Population mean (known)
population_mean = 16

# Z-score calculation
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# P-value calculation
p_value = stats.norm.sf(abs(z_score)) * 2  # two-tailed test

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt

def plot_errors():
    # Define the distribution for Type 1 and Type 2 errors
    population_mean = 0
    sample_mean = 1
    std_dev = 1
    sample_size = 30

    # Define decision thresholds
    alpha = 0.05
    critical_value = stats.norm.ppf(1 - alpha / 2)

    # Visualize the two distributions
    x = np.linspace(-4, 4, 1000)
    pop_dist = stats.norm.pdf(x, population_mean, std_dev)
    sample_dist = stats.norm.pdf(x, sample_mean, std_dev / np.sqrt(sample_size))

    plt.plot(x, pop_dist, label='Null Distribution (Population Mean)')
    plt.plot(x, sample_dist, label='Alternative Distribution (Sample Mean)')

    # Type I and Type II errors visualization
    plt.fill_between(x, 0, pop_dist, where=(x < -critical_value) | (x > critical_value), color='red', alpha=0.5, label='Type I Error')
    plt.fill_between(x, 0, sample_dist, where=(x > critical_value), color='green', alpha=0.5, label='Type II Error')

    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.show()

plot_errors()
```

```python
#9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, sample_size - 1)

    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    return sample_mean - margin_of_error, sample_mean + margin_of_error

# Sample data
data = np.random.normal(loc=20, scale=5, size=100)

# Confidence interval calculation
lower, upper = confidence_interval(data)
print(f"95% Confidence Interval: ({lower}, {upper})")
```

```python
#16)  Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
from scipy import stats

# Simulate two random samples with different variances
sample1 = np.random.normal(loc=10, scale=2, size=30)  # Sample 1 with variance 4 (scale=2)
sample2 = np.random.normal(loc=10, scale=4, size=30)  # Sample 2 with variance 16 (scale=4)

# Perform F-test
f_stat, p_value = stats.levene(sample1, sample2)  # Levene's test to compare variances

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances of the two samples are not significantly different.")
```

```python
#19) Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
import scipy.stats as stats

def check_anova_assumptions(groups):
    # Normality: Check using Shapiro-Wilk test
    for i, group in enumerate(groups, start=1):
        stat, p = stats.shapiro(group)
        print(f"Group {i} Normality Test (Shapiro-Wilk): Statistic = {stat}, P-value = {p}")
        if p < 0.05:
            print(f"Group {i}: Data is not normally distributed")
        else:
            print(f"Group {i}: Data is normally distributed")

    # Homogeneity of variance: Levene's Test
    stat, p_value = stats.levene(*groups)
    print(f"\nLevene's Test for Homogeneity of Variance: Statistic = {stat}, P-value = {p_value}")
    if p_value < 0.05:
        print("The variances are significantly different.")
    else:
        print("The variances are not significantly different.")

# Check assumptions for the three groups in ANOVA
check_anova_assumptions([group1, group2, group3])
```

```python
#23)  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
# Simulate data
sample_data = np.random.normal(loc=50, scale=10, size=100)

# Perform one-sample Z-test (population mean = 50)
population_mean = 50
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
sample_std = np.std(sample_data, ddof=1)

z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = stats.norm.sf(abs(z_score)) * 2  # two-tailed test

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

```python
#24) Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
# Sample data
sample_data = np.random.normal(10, 3, 30)

# Population variance (known)
population_variance = 9  # variance = standard deviation squared

# Chi-square test for variance
sample_size = len(sample_data)
sample_variance = np.var(sample_data, ddof=1)
chi_square_stat = (sample_size - 1) * sample_variance / population_variance

# P-value for chi-square test
p_value = 1 - stats.chi2.cdf(chi_square_stat, df=sample_size - 1)

print(f"Chi-square statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

```python
#25) Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
# Sample data for two groups (successes and trials)
group1_successes = 45
group1_trials = 100
group2_successes = 30
group2_trials = 100

# Proportions
p1 = group1_successes / group1_trials
p2 = group2_successes / group2_trials

# Pooled proportion
pooled_p = (group1_successes + group2_successes) / (group1_trials + group2_trials)

# Z-score calculation
z_score = (p1 - p2) / np.sqrt(pooled_p * (1 - pooled_p) * (1/group1_trials + 1/group2_trials))
p_value = stats.norm.sf(abs(z_score)) * 2  # two-tailed test

# Output
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

```python
#26) Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate two random samples with different variances
np.random.seed(0)
sample1 = np.random.normal(loc=10, scale=2, size=50)  # Sample 1 with standard deviation of 2
sample2 = np.random.normal(loc=10, scale=5, size=50)  # Sample 2 with standard deviation of 5

# Calculate the sample variances
var_sample1 = np.var(sample1, ddof=1)  # Sample variance for sample1
var_sample2 = np.var(sample2, ddof=1)  # Sample variance for sample2

# Perform F-test
f_stat = var_sample1 / var_sample2  # F-statistic is the ratio of variances
df1 = len(sample1) - 1  # Degrees of freedom for sample1
df2 = len(sample2) - 1  # Degrees of freedom for sample2

# Calculate the p-value using the F-distribution
p_value = stats.f.sf(f_stat, df1, df2)

# Output results
print(f"Variance of Sample 1: {var_sample1}")
print(f"Variance of Sample 2: {var_sample2}")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances of the two samples are not significantly different.")

# Visualization: Compare the distributions of both samples with boxplots
plt.figure(figsize=(8, 6))
plt.boxplot([sample1, sample2], labels=["Sample 1", "Sample 2"])
plt.title("F-test: Comparison of Variances between Sample 1 and Sample 2")
plt.ylabel("Values")
plt.show()
```

```python
#27) Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate observed frequencies for 6 categories (e.g., rolling a die)
np.random.seed(0)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=100, p=[0.1, 0.2, 0.2, 0.2, 0.15, 0.15])

# Calculate observed frequencies
observed_frequencies = [np.sum(observed == i) for i in range(1, 7)]
print(f"Observed Frequencies: {observed_frequencies}")

# Define the expected frequencies for a fair die (uniform distribution)
expected_frequencies = np.array([100/6] * 6)

# Perform the Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# Output the results
print(f"\nChi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies do not match the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies match the expected frequencies.")

# Visualization: Plot the observed and expected frequencies
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))

plt.figure(figsize=(8, 6))
plt.bar(x - 0.2, observed_frequencies, width=0.4, label='Observed', color='blue', alpha=0.7)
plt.bar(x + 0.2, expected_frequencies, width=0.4, label='Expected', color='orange', alpha=0.7)

plt.xticks(x, labels)
plt.xlabel('Categories (Die Rolls)')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness of Fit: Observed vs Expected Frequencies')
plt.legend()
plt.show()
```
